```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
# loading dataset to a pandas DataFrame
data=pd.read_csv('creditcard.csv')
```

```python
df=data
```

```python
# print top 5 rows of the dataset
df.head()
```

```python
# print bottom 5 rows of the dataset
df.tail()
```

```python
# dataset information

df.info()
```

```python
# checking the missing value in each column

df.isnull().sum()
```

```python
# distribution of legit transaction & fraudulent transaction
df['Class'].value_counts()
```

This dataset is highly unbalanced


0 --> Normal transaction


1 --> fraudulent transaction

```python
# Seprating the data analysis

legit = df[data.Class == 0]
fraud = df[data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
# statistical measures of the data

legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
# compare the valuess for both transactions

data.groupby('Class').mean()
```

Under-sampling

Build a sample dataset containing similar distribution of normal transactions

Number of Fraudulent Transactions-->86

```python
legit_sample = legit.sample(n=86)
```

concatenating two DataFrames

```python
new_dataset = pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

spliting the data into features & Targets

```python
X = new_dataset.drop(columns='Class',axis=1)
Y = new_dataset['Class']
```

```python
print(X)
```

```python
print(Y)
```

Split the data into Training data & Testing data

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y , test_size=0.2 , stratify=Y , random_state=42)
```

```python
print(X.shape,X_train.shape , X_test.shape)
```

Model Training

Logistic Regression

```python
model = LogisticRegression(solver ='lbfgs', max_iter=1000)
```

```python
# training the Logistic Regression Model with Training Data
model.fit(X_train , Y_train)
```

Model Evaluation

Accuracy score

```python
# accuracy on training data
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,Y_train)
```

```python
print('Accuracy on Training data:',training_data_accuracy)
```

```python
# accuracy on test data
X_test_prediction = model.predict(X_test)
test_data_accuracy = accuracy_score(X_test_prediction,Y_test)
```

```python
print('Accuracy on test data:', test_data_accuracy)
```
